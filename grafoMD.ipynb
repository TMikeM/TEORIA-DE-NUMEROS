```python
import networkx as nx
import networkx.algorithms.isomorphism as iso

import scipy.io as sio
import numpy as np
```

```python
'''
---------------- NOMENCLATURA DE LOS EDIFICIOS: -------------------------|
- EN: Entrada de la 27         | - FINGF: Facultad de Fisicomecanicas
- CH: Ciencias Humanas         | - DI: Dise√±o industrial
- IM: Ingenieria Mecanica      | - CEIM:              
- CT: Camilo Torres            | - CAYFAV: CapriuisyFavuis
- CEN: Centic                  | - LABAT: Laboratorio de alta tension
- FDC : Facultad de Ciencias   | - LABDI: Laboratorio de diseli undustrial
- IE: Ingenieria Electrica     | - LABHI: Laboratorio de Hidraulica
- II: Ingenieria Industrial    | - PDA: Planta de Aceros
- IL: Instituto de Lenguas     | - JBV: Jorge Bautista V
- IQ: Ingenieria Quimica       | - DACA: Daniel Casas
- AMM: Aula Maxima de Mecanica | - AMF: Aula Maxima de Fisica
-E3T: Electrica Electronica y Telecomunicaciones 

'''
```

```python
#Creacion del Grafo 
G = nx.Graph() #creacion del grafo

#Creacion del arreglo de vertices y agregarlos al nodo.
nodes=["EN","CH","IM","CT","CEN","FDC","II","IL","IQ"
       ,"FINGF","DI","CEIAM","CAYFA","LABAT","TDI","LABHI"
       ,"PDA","JBV","DACA","AMM","AMF","E3T"]

nodes2=["I0","I1","I2","I3","I4","I5","I6","I7","I8","I9"] #intersecciones

G.add_nodes_from(nodes) #Agregando los vertices/nodos del grafo
G.add_nodes_from(nodes2) #Agregando las intesecciones del grafo
```

```python
#Imprime los vertices en pantalla
print(G.nodes)
```

```python
#Creacion de las relaciones entre las intersecciones y las aristas del grafo


#Creacion de las aristas entre edificios
edges= [("EN","I0"),("IL","IO"),("IL","I1"),("AMM","I1"),("IM","I1"),
        ("CT","I2"),("CEN","I2"),("CEN","I3"),("FDC","I3"),("CT","I3"),
        ("IQ","I3"),("AMF","I3"),("CAYFA","I3"),("JBV","I4"),("PDA","I4"),
        ("FINGF","I4"),("CEIAM","I5"),("AMF","I8"),("CAYFA","I8"),("LABAT","I8"),
        ("LABHI","I8"),("TDI","I8"),("IL","I6"),("DI","I6"),("DI","I7"),
        ("E3T","I7"),("II","I7"),("EN","I9"),("CH","I9"),("IM","I9"),("DACA","I5")]

# Creacion de aristas entre las intersecciones
edges2=[("I0","I1"),("I1","I2"),("I2","I6"),("I2","I9"),("I2","I3"),
        ("I3","I4"),("I4","I5"),("I6","I7"),("I7","I8")] 

G.add_edges_from(edges) 
G.add_edges_from(edges2) 
print(G.edges)
```

```python
nx.draw(G, with_labels=True)
```
